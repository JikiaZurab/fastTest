```python
import pandas as pd
```

```python
before = pd.read_csv("student_before_score.csv")
before
```

```python
after = pd.read_csv("student_after_score.csv")
after
```

```python
group = pd.read_csv("student_group.csv")
group
```

```python
merged_scores = pd.merge(before, after, on = "student_id", how = "inner")
merged_scores
```

```python
full_table = pd.merge(merged_scores, group, on = "student_id", how = "inner")
full_table
```

```python
full_table["Differece"] = full_table["after_score"] - full_table["before_score"]
full_table
```

```python
for group in full_table["group"].unique():
    mean_before = full_table.loc[full_table["group"] == group, "before_score"].mean()
    mean_after = full_table.loc[full_table["group"] == group, "after_score"].mean()
    print(f"Group with {group} had before {mean_before:.4} and after {mean_after:.4}")
```
